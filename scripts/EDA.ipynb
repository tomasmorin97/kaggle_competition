```python
# Definimos una funci√≥n que haga que las variables binarias tengan 0 o 1:
def binary(dataset, column):
    dataset = pd.get_dummies(dataset, columns=[column], drop_first=True, dtype=int)
    return dataset
```

```python
# Aplicar la funci√≥n a las columnas correspondientes
df_train = binary(df_train, 'Street')
df_train = binary(df_train, 'CentralAir')
df_train = binary(df_train, 'Alley')
df_train = binary(df_train, 'RoofStyle')
df_train = binary(df_train, 'RoofMatl')
df_train = binary(df_train, 'MasVnrType')
df_train = binary(df_train, 'Heating')
df_train = binary(df_train, 'GarageType')
df_train = binary(df_train, 'SaleType')
df_train = binary(df_train, 'SaleCondition')
```

```python
# Definir la funci√≥n para mapear valores a valores num√©ricos
def mapear_a_valor_numerico(columna, mapeo_manual):
    # Aplicar el mapeo manual a la columna
    columna_numerica = columna.map(mapeo_manual)
    return columna_numerica

# Nombre de la columna que deseas modificar
columna_a_modificar = 'PoolQC'

# Definir el mapeo manual proporcionado por el usuario
mapeo_manual = {'NA': 0, 'Fa': 1, 'TA': 2, 'Gd': 3, 'Ex': 4}

# Aplicar la funci√≥n a la columna espec√≠fica
df_train[columna_a_modificar] = mapear_a_valor_numerico(df_train[columna_a_modificar], mapeo_manual)

# Despu√©s de cambiar los valores
df_train['PoolQC']
```
