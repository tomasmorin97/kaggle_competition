```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# Crear un boxplot utilizando Seaborn
sns.set(style="whitegrid")  # Puedes personalizar el estilo según tus preferencias

# Boxplot de varias variables
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['YearBuilt']], palette="Set3")  # Selecciona un esquema de colores según tus preferencias

# Añadir título y etiquetas
plt.title("Boxplot de Varias Variables")
plt.xlabel("Variables")
plt.ylabel("Valores")

# Mostrar el boxplot
plt.show()
```
